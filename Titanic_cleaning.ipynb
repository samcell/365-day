```python
import pandas as pd
pd.set_option('display.max_colwidth',None)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df_copy = df
```

```python
df_copy = df_copy.drop(columns= ['Cabin'])
```

```python
df_copy.head()
```

```python
df_copy.drop_duplicates(inplace=True)
```

```python
df_copy
```

```python
df_copy['Sex'] = df_copy['Sex'].str.capitalize()
df_copy.head(10)
```

```python
# 1. Remove anything inside parentheses ( ... )
df_copy['Name'] = df_copy['Name'].str.replace(r"\(.*?\)", "", regex=True)

# 2. Remove anything inside quotation marks " ... "
df_copy['Name'] = df_copy['Name'].str.replace(r"\".*?\"", "", regex=True)

# 3. Remove leading/trailing whitespaces
df_copy['Name'] = df_copy['Name'].str.strip()

# 4. Remove double spaces that may appear after cleaning
df_copy['Name'] = df_copy['Name'].str.replace(r"\s+", " ", regex=True)
df_copy
```

```python
# Step 1: Extract Last Name
df_copy['Last_Name'] = df_copy['Name'].str.extract(r'^([^,]+)')

# Step 2: Extract Title
df_copy['Title'] = df_copy['Name'].str.extract(r',\s*([^\.]+)\.')

# Step 3: Extract First Name (everything after the title)
df_copy['First_Name'] = df_copy['Name'].str.extract(r'\.\s*(.*)')

# Step 4: Strip whitespace
df_copy['Title'] = df_copy['Title'].str.strip()
df_copy['First_Name'] = df_copy['First_Name'].str.strip()
df_copy
```

```python
df_copy.info()
```

```python
df_copy['Age'].fillna(df_copy['Age'].median(),inplace= True)
df_copy['Age'] = df_copy['Age'].astype(int)
```
