```python
import random
```

```python
Name = input("What's your name?\n")
```

```python
print(f"Welcome, {Name} to the number guessing game.\nYou have 5 chances to guess the right number")
```

```python
low_number = int(input("Enter the low number:"))
high_number = int(input("Enter the high number:"))
```

```python
Number = random.randint(low_number,high_number)
chances = 5
remaining_chance = 0
```

```python
while remaining_chance < chances:
    remaining_chance += 1
    remainder = chances - remaining_chance

    try:
        Guess = int(input("Enter your guess: "))
    except ValueError:
        print(f"Invalid input! Please enter a number. You still have {remainder} attempts left.")
        continue  # go back to the top of the loop and ask again

    if Guess == Number:
        print(f'Correct! The number is {Number}. You guessed it in {remaining_chance} attempts.')
        break
    elif remaining_chance >= chances:
        print(f'Sorry! The number was {Number}. Better luck next time.')
    elif Guess < Number:
        print(f'Number is low. Try again - You have {remainder} attempts')
    elif Guess > Number:
        print(f'Number is high. Try again - You have {remainder} attempts')
```
