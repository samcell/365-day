```python
import random
```

```python
Name = input("What's your name?\n")
```

```python
print(f"Welcome, {Name} to the number guessing game.\nYou have 5 chances to guess the right number")
```

```python
low_number = int(input("Enter the low number:"))
high_number = int(input("Enter the high number:"))
```

```python
Number = random.randint(low_number,high_number)
chances = 5
remaining_chance = 0
```

```python
while remaining_chance < chances:
  remaining_chance = remaining_chance + 1
  Guess = int(input("Enter your guess: "))

if Guess == Number:
  print(f'Correct! The number is {Number}. You guessed it in {remaining_chance} attempts.')
  break
elif Guess != Number and remaning_chance >= chances:
  print(f'Sorry! The number was {Number}. Better luck next time.')
```
